02 ``Figure``对象
在matplotlib中，``Figure``对象是绘图的基本单元。它包含``Axes``对象，每个``Axes``对象代表一个坐标系，用于绘制图形。

``Figure``对象有很多属性和方法，包括：

- ``add_subplot``：添加子图
- ``add_axes``：添加坐标系
- ``gca``：获取当前坐标系
- ``sca``：设置当前坐标系
- ``delaxes``：删除坐标系
- ``suptitle``：设置标题
- ``text``：添加文本
- ``legend``：添加图例
- ``colorbar``：添加颜色条
- ``savefig``：保存图片
- ``show``：显示图片

````python
plt.Figure(figsize=(10, 8))
plt.plot(6, 3)
plt.plot(3, 3 * 2)
plt.show()
````

04 设置``label``和``legend``
设置``label``和``legend``是为了方便在图中标识不同的曲线，并在图中添加图例。

````python
x = [1, 2, 3, 4, 5]      # x轴数据                          
y1 = [2, 4, 1, 3, 5]     # 第一个y轴数据
y2 = [3, 2, 5, 1, 4]     # 第二个y轴数据

plt.plot(x, y1, label='y1')   # 画第一个曲线，并设置label为'y1'
plt.plot(x, y2, label='y2')   # 画第二个曲线，并设置label为'y2'

plt.xlabel('x')   # 设置x轴标签
plt.ylabel('y')   # 设置y轴标签

plt.title('My Graph')   # 设置图标题

plt.legend()   # 显示图例

plt.show()   # 显示图形
````

05 添加注释

有时候需要对特定的点进行标注，可以使用plt.annotate()函数。

```python
plt.annotate(s='注释文字', xy=(x, y), xytext=(x+10, y+10), arrowprops=dict(facecolor='black', shrink=0.05))
```

参数说明：

s：注释文字

xy：注释的位置           

xytext：注释文字的位置

arrowprops：箭头属性，包括facecolor：箭头颜色，shrink：箭头缩短比例     

````python
import numpy as np
x = np.linspace(0, 10, 200) # 生成100个等差数列
y = np.sin(x) # 计算正弦值
plt.plot(x, y, linestyle=':', color='r', marker='o', markersize=5, markerfacecolor='b', markeredgecolor='g')   # 绘制正弦曲线
plt.annotate('标记点', xy=(3, np.sin(3)), xytext=(4, -0.5), arrowprops=dict(facecolor='black', shrink=0.05)) # 标注sin(x)
plt.show()
````

05 使用子图
如果需要将多张子图展示在一起，可以使用 ``subplot()``实现。即在调用 ``plot()``函数之前需要先调用subplot()函数，并指定行数和列数。

``subplot(nrows, ncols, index)``

- nrows: 行数
- ncols: 列数
- index: 子图的索引，从1开始，从左到右，从上到下排列。

例如，要在一张图中绘制两个子图，第一张子图为直方图，第二张子图为散点图，可以用以下代码：

````python
import matplotlib.pyplot as plt
import numpy as np

# 产生数据
x = np.random.randn(1000)
y = np.random.randn(1000)

# 创建子图
plt.subplot(2, 1, 1)
plt.hist(x, bins=20, color='g')

plt.subplot(2, 1, 2)
plt.scatter(x, y, color='r')

# 显示图形
plt.show()
````

# 01 bar
**柱状图**
生成一个单系列的柱状图比较简单，只要确定x轴和y轴的数据，利用bar()函数就能生成。

````python
x = np.arange(10) # 生成10个数字的数组
y = np.random.randint(0, 20, 10) # 生成10个随机整数
plt.bar(x, y) # 绘制条形图
plt.show() # 显示图形
````

除了单系列柱状图，matplotlib还提供了多系列柱状图，堆叠图，水平向的条纹图等。plt.plot()适用于基本图表的绘制，kind可选类型有线形图，柱状图，密度图，堆叠图，面积图等，以横纵坐标两个维度为主。grid是否显示网格，alpha透明度，color颜色，label图例标签，linewidth线宽，marker标记，markersize标记大小，markevery标记位置，fillstyle填充样式等参数可选。

````python
# 多系列柱状图
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C'])
df.plot(kind='bar', grid=True, colormap='summer_r') # colormap='summer_r'为设置颜色风格
````

想要实现堆叠效果，一定要加上stacked=true，否则输出图形就是一般的柱状图。

````python
# 多系列堆叠图
df.plot(kind='bar', grid=True, colormap='Blues_r', stacked=True)
````

水平向的条状图调用的是barh()函数，垂直向的条状图调用的是bar()函数。

````python
# 水平向
df.plot.barh(grid=True, colormap='BuGn_r')
````

与垂直柱状图一样，如果想要柱状图实现堆叠效果，则加上:stacked=True参数即可。

````python
df.plot.barh(grid=True, colormap='BuGn_r',stacked=True)
````

# 03 pie
**饼图**

````python
# matplotlib 中饼图的实现是用pie()函数，必须输入的参数是饼图每个部分的值

x = np.random.randint(1, 10, 3)
plt.pie(x)
plt.show()
````

pie()函数参数说明：

- labels：饼图的标签，列表或元组类型，默认为None，即没有标签
- explode：饼图的突出显示，列表或元组类型，默认为None，即没有突出显示
- autopct：饼图的百分比格式，字符串或函数类型，默认为None，即没有百分比格式
- shadow：饼图是否有阴影，布尔类型，默认为False，即没有阴影
- startangle：饼图的起始角度，浮点数类型，默认为0，即从0度开始
- radius：饼图的半径，浮点数类型，默认为1，即饼图的直径为1
- counterclock：饼图的方向，布尔类型，默认为False，即顺时针方向
- wedgeprops：饼图的属性，字典类型，默认为None，即没有属性
- textprops：饼图标签的属性，字典类型，默认为None，即没有属性
- center：饼图的中心，元组类型，默认为None，即饼图的中心在图形的中心
- frame：饼图的边框，布尔类型，默认为False，即没有边框

````python
sizes = [2, 5, 12]
labels = ['A', 'B', 'C']
plt.subplot(1, 2, 1)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=100,shadow=False)

plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0,shadow=True)

plt.show()
````

# 04 直方图hist
直方图绘制为hist()函数，参数如下：
- data：绘制的直方图数据，可以是一维数组或二维数组。
- bins:直方图的长条形数目，可选项，默认为10。
- range：直方图的范围，可选项，默认为None，表示自动选择。
- density：是否将直方图显示为概率密度直方图，可选项，默认为False。
- cumulative：是否绘制累积直方图，可选项，默认为False。
- align：直方图的对齐方式，可选项，默认为'mid',表示将每个长条形的中心对齐。
- orientation：直方图的方向，可选项，默认为'vertical',表示垂直方向。
- color：直方图的颜色，可选项，默认为'b',表示蓝色。
- alpha：直方图的透明度，可选项，默认为1.0。
- label：直方图的标签，可选项，默认为None。
- stacked：是否将多个直方图堆叠显示，可选项，默认为False。
- normed：是否将直方图显示为概率密度直方图，可选项，默认为False。
- Returns：返回一个直方图对象。

````python
s = pd.Series(np.random.randn(1000))
s.hist(bins=20,histtype='bar',alpha=0.3,color='steelblue',edgecolor='none')
````

````python
# 密度图
s.plot(kind='kde',style='-',color='r')
````

# 05 polar
极坐标图
matplotlib的pyplot字库提供了绘制极坐标图的方法，在调用subplot()创建子图时通过设置projection参数为polar即可。

````python
ax1 = plt.subplot(121, projection='polar')
````

部分参数意义：
- theta：角度数据
- radii：极径数据
- theta_direction：极坐标轴方向，'counterclockwise'表示逆时针方向，'clockwise'表示顺时针方向
- width：极径线的宽度
- color：极径线的颜色
- edgecolor：极径线的边框颜色
- linestyle：极径线的线型
- theta_zero_location：极坐标轴的起点，'N'表示极坐标轴的正上方，'NW'表示极坐标轴的左上方，'W'表示极坐标轴的左方，'SW'表示极坐标轴的左下方，'S'表示极坐标轴的正下方，'SE'表示极坐标轴的右下方，'E'表示极坐标轴的右方，'NE'表示极坐标轴的右上方
- thetagrids: 极坐标网格线的角度值，默认值为None，表示不显示网格线
- theta_offset: 极坐标轴的偏移量，默认值为0，表示不偏移
- rgrids: 极径网格线的极径值，默认值为None，表示不显示网格线
- rmax: 极径轴的最大值，默认值为None，表示自动计算

````python
N = 20
theta = np.linspace(0, 2*np.pi, N, endpoint=False) # 角度
radii = 10*np.random.rand(N) # 半径
width = np.pi/4*np.random.rand(N) # 宽度
ax = plt.subplot(111, polar=True) # 极坐标
bars = ax.bar(theta, radii, width=width, bottom=0.0)
````

````python
theta = np.arange(0, 2*np.pi, np.pi/4) # 数据角度
r = np.arange(1, 9, 1) # 数据极径
area = 100*np.arange(1, 9, 1) # 数据三点面积
colors = theta
ax2 = plt.subplot(122, projection='polar')
ax2.scatter(theta, r, s=area, c=colors, alpha=0.5,cmap='hsv')
````

# 06 boxplot
**箱型图**
绘制箱型图，用plt.boxplot()函数。箱型图是利用数据中的五个统计量：最小值、第一四分位数、中位数、第三四分位数、最大值，将数据分为上下两部分，并用矩形条表示。箱型图可以直观地反映出数据分布的形态、位置、离散程度、峰度等。
它可以粗略地看出数据是否具有有对称性，分布的分散程度等信息，特别可以用于对几个样本的比较。


箱型图各部分含义：
- 左侧：最小值
- 右侧：最大值
- 顶部：第一四分位数
- 底部：第三四分位数
- 中间：中位数

boxplot()函数的语法：

```python
plt.boxplot(data, notch=False, sym='b+', vert=True, whis=1.5, positions=None, widths=None, patch_artist=False, bootstrap=None, usermedians=None, conf_intervals=None, meanline=False, showmeans=False, showcaps=True, showbox=True, showfliers=True, boxprops=None, labels=None, flierprops=None, medianprops=None, meanprops=None, capprops=None, whiskerprops=None, manage_xticks=True, autorange=False, zorder=None, precision=0.01, ax=None)
```

参数说明：
- data：数据，可以是一维数组或二维数组。
- notch：是否显示不规则的箱体。
- sym：箱体的形状，默认是'b+'，表示实心的蓝色加号。
- vert：是否垂直显示箱体。
- whis：设置四分位距，默认为1.5。
- positions：箱体的位置，默认为None，表示自动分配位置。
- widths：箱体的宽度，默认为0.5。
- patch_artist：是否使用PathCollection来绘制箱体。
- bootstrap：是否使用bootstrap方法计算统计量。
- usermedians：自定义中位数。
- conf_intervals：置信区间。
- meanline：是否显示平均线。
- showmeans：是否显示中位数。
- showcaps：是否显示顶部和底部的线。
- showbox：是否显示箱体。
- showfliers：是否显示异常值。
- boxprops：箱体的属性。
- labels：箱体的标签。
- flierprops：异常值属性。
- medianprops：中位数属性。
- meanprops：平均线属性。
- capprops：顶部和底部线属性。
- whiskerprops：四分位线属性。
- manage_xticks：是否自动管理x轴刻度。
- autorange：是否自动调整y轴范围。
- zorder：图层顺序。
- precision：计算置信区间的精度。
- ax：指定绘图区域。

````python
np.random.seed(100) # 生成随机数
data = np.random.normal(size=1000, loc=0, scale=1)
plt.boxplot(data, sym='o',whis=1.5)
plt.show()
````

````python
# 多箱型图
np.random.seed(100) # 生成随机数
data = np.random.normal(size=(1000, 4),loc=0,scale=1) # 生成1000组4个数据
labels = ['A', 'B', 'C', 'D']
plt.boxplot(data, labels=labels) # 绘制箱型图
plt.show()
````

箱型图也可以是横向的，加上 vert = False 参数即可。

````python
plt.boxplot(data, labels=labels,vert = False)
plt.show() 
````

# 可视化控制

前面用matplotlib绘图的时候，我们已经知道了如何绘制折线图、散点图、直方图、饼图等图表。这些图表都是由不同类型的图元组成的，比如折线图由折线、点、标签组成，散点图由点、线、标签组成，直方图由柱状图、刻度线、标签组成，饼图由扇形、标签组成。

在绘图的时候，我们可以控制图表的颜色、线型、透明度、大小、标签、标题、网格、轴线等。下面我们来看看这些控制的具体方法。

## 颜色

matplotlib中颜色的控制主要通过颜色名称、RGB值、十六进制值来实现。

### 颜色表示
```
b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white

```

````python
# 先用numpy生成四条线，再对四条线的颜色进行设置：

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
y = np.arange(1, 5)
````
