02 ``Figure``对象
在matplotlib中，``Figure``对象是绘图的基本单元。它包含``Axes``对象，每个``Axes``对象代表一个坐标系，用于绘制图形。

``Figure``对象有很多属性和方法，包括：

- ``add_subplot``：添加子图
- ``add_axes``：添加坐标系
- ``gca``：获取当前坐标系
- ``sca``：设置当前坐标系
- ``delaxes``：删除坐标系
- ``suptitle``：设置标题
- ``text``：添加文本
- ``legend``：添加图例
- ``colorbar``：添加颜色条
- ``savefig``：保存图片
- ``show``：显示图片

````python
plt.Figure(figsize=(10, 8))
plt.plot(6, 3)
plt.plot(3, 3 * 2)
plt.show()
````

03 坐标轴及标签

````python
plt.xlim(0, 6) # x轴坐标轴
plt.ylim((0, 3)) # y轴坐标轴
plt.xlabel('X') # x轴标签
plt.ylabel('Y') # y轴标签
plt.show() # 显示图形
````

04 设置``label``和``legend``
设置``label``和``legend``是为了方便在图中标识不同的曲线，并在图中添加图例。

````python
x = [1, 2, 3, 4, 5]      # x轴数据                          
y1 = [2, 4, 1, 3, 5]     # 第一个y轴数据
y2 = [3, 2, 5, 1, 4]     # 第二个y轴数据

plt.plot(x, y1, label='y1')   # 画第一个曲线，并设置label为'y1'
plt.plot(x, y2, label='y2')   # 画第二个曲线，并设置label为'y2'

plt.xlabel('x')   # 设置x轴标签
plt.ylabel('y')   # 设置y轴标签

plt.title('My Graph')   # 设置图标题

plt.legend()   # 显示图例

plt.show()   # 显示图形
````

````python
x = [1, 2, 3, 4, 5]      # x轴数据                          
y1 = [2, 4, 1, 3, 5]     # 第一个y轴数据
y2 = [3, 2, 5, 1, 4]     # 第二个y轴数据

plt.plot(x, y1, label='y1')   # 画第一个曲线，并设置label为'y1'
plt.plot(x, y2, label='y2')   # 画第二个曲线，并设置label为'y2'

plt.xlabel('x')   # 设置x轴标签
plt.ylabel('y')   # 设置y轴标签

plt.title('My Graph')   # 设置图标题

plt.legend(loc='upper right')   # 设置图例位置

plt.show() 
````

05 添加注释

有时候需要对特定的点进行标注，可以使用plt.annotate()函数。

```python
plt.annotate(s='注释文字', xy=(x, y), xytext=(x+10, y+10), arrowprops=dict(facecolor='black', shrink=0.05))
```

参数说明：

s：注释文字

xy：注释的位置           

xytext：注释文字的位置

arrowprops：箭头属性，包括facecolor：箭头颜色，shrink：箭头缩短比例     

````python
import numpy as np
x = np.linspace(0, 10, 200) # 生成100个等差数列
y = np.sin(x) # 计算正弦值
plt.plot(x, y, linestyle=':', color='r', marker='o', markersize=5, markerfacecolor='b', markeredgecolor='g')   # 绘制正弦曲线
plt.annotate('标记点', xy=(3, np.sin(3)), xytext=(4, -0.5), arrowprops=dict(facecolor='black', shrink=0.05)) # 标注sin(x)
plt.show()
````

05 使用子图
如果需要将多张子图展示在一起，可以使用 ``subplot()``实现。即在调用 ``plot()``函数之前需要先调用subplot()函数，并指定行数和列数。

``subplot(nrows, ncols, index)``

- nrows: 行数
- ncols: 列数
- index: 子图的索引，从1开始，从左到右，从上到下排列。

例如，要在一张图中绘制两个子图，第一张子图为直方图，第二张子图为散点图，可以用以下代码：

````python
import matplotlib.pyplot as plt
import numpy as np

# 产生数据
x = np.random.randn(1000)
y = np.random.randn(1000)

# 创建子图
plt.subplot(2, 1, 1)
plt.hist(x, bins=20, color='g')

plt.subplot(2, 1, 2)
plt.scatter(x, y, color='r')

# 显示图形
plt.show()
````

# 01 bar
**柱状图**
生成一个单系列的柱状图比较简单，只要确定x轴和y轴的数据，利用bar()函数就能生成。

````python
x = np.arange(10) # 生成10个数字的数组
y = np.random.randint(0, 20, 10) # 生成10个随机整数
plt.bar(x, y) # 绘制条形图
plt.show() # 显示图形
````

除了单系列柱状图，matplotlib还提供了多系列柱状图，堆叠图，水平向的条纹图等。plt.plot()适用于基本图表的绘制，kind可选类型有线形图，柱状图，密度图，堆叠图，面积图等，以横纵坐标两个维度为主。grid是否显示网格，alpha透明度，color颜色，label图例标签，linewidth线宽，marker标记，markersize标记大小，markevery标记位置，fillstyle填充样式等参数可选。

````python
# 多系列柱状图
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C'])
df.plot(kind='bar', grid=True, colormap='summer_r') # colormap='summer_r'为设置颜色风格
````

想要实现堆叠效果，一定要加上stacked=true，否则输出图形就是一般的柱状图。

````python
# 多系列堆叠图
df.plot(kind='bar', grid=True, colormap='Blues_r', stacked=True)
````

水平向的条状图调用的是barh()函数，垂直向的条状图调用的是bar()函数。

````python
# 水平向
df.plot.barh(grid=True, colormap='BuGn_r')
````

与垂直柱状图一样，如果想要柱状图实现堆叠效果，则加上:stacked=True参数即可。

````python
df.plot.barh(grid=True, colormap='BuGn_r',stacked=True)
````

# 02 scatter
**散点图**
绘制散点图，主要用到plt.scatter()函数。

scatter函数参数说明：

- x：x轴数据
- y：y轴数据
- s：点的大小，控制点的大小，默认为20
- c：点的颜色
- marker：点的形状，指定散点图点的形状，默认为圆形
- alpha：透明度
- cmap：颜色映射
- norm：颜色映射的范围
- vmin、vmax：颜色映射的最小最大值
- edgecolors：边框颜色
- linewidths：边框宽度
- label：图例标签

````python
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x, y)
plt.show()
````

进行简单的一些参数设置:

````python
x = np.random.rand(50) # generate 50 random numbers
y = np.random.rand(50)
colors = np.random.rand(50) # generate 50 random colors
sizes = (30 * np.random.rand(50))**2
plt.scatter(x, y, c=colors, s=sizes,alpha=0.5)
plt.show()
````

散点矩阵图 scatter_matrix，diagonal=''为每个指标的频率图，有kde及hist两个参数可选；range_padding是图像在x轴，y轴远点附近的留白，值越大，图像离坐标原点的距离越大。

````python
import pandas as pd
from pandas.plotting import scatter_matrix
df = pd.DataFrame(np.random.randn(100, 4), columns = list('abcd'))
pd.scatter_matrix(df, figsize = (8,6),marker = 'o',diagonal = 'kde',alpha = 0.4,range_padding = 0.05)
````

# 03 pie
**饼图**

````python
# matplotlib 中饼图的实现是用pie()函数，必须输入的参数是饼图每个部分的值

x = np.random.randint(1, 10, 3)
plt.pie(x)
plt.show()
````

pie()函数参数说明：

- labels：饼图的标签，列表或元组类型，默认为None，即没有标签
- explode：饼图的突出显示，列表或元组类型，默认为None，即没有突出显示
- autopct：饼图的百分比格式，字符串或函数类型，默认为None，即没有百分比格式
- shadow：饼图是否有阴影，布尔类型，默认为False，即没有阴影
- startangle：饼图的起始角度，浮点数类型，默认为0，即从0度开始
- radius：饼图的半径，浮点数类型，默认为1，即饼图的直径为1
- counterclock：饼图的方向，布尔类型，默认为False，即顺时针方向
- wedgeprops：饼图的属性，字典类型，默认为None，即没有属性
- textprops：饼图标签的属性，字典类型，默认为None，即没有属性
- center：饼图的中心，元组类型，默认为None，即饼图的中心在图形的中心
- frame：饼图的边框，布尔类型，默认为False，即没有边框

````python
sizes = [2, 5, 12]
labels = ['A', 'B', 'C']
plt.subplot(1, 2, 1)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=100,shadow=False)

plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0,shadow=True)

plt.show()
````
